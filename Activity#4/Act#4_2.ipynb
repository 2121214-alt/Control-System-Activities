```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1] #numerator
den = [1, 2, 2] #denominator s^2 + 2s + 2

G = ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zeros = ctrl.zeros(G)
print("Zeros:", zeros)
```

```python
ctrl.pzmap(G, grid=True,)
plt.show()
```
